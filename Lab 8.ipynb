```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# 1) FoodTruck.csv question

```python
df = pd.read_csv("Foodtruck.csv")
df
```

```python
feature = df['Population'].values
label = df['Profit'].values
```

```python
feature_train,feature_test,label_train,label_test = train_test_split(feature,label,test_size=0.2,random_state=42)
```

```python
regressor = LinearRegression()
```

```python
feature_train = feature_train.reshape(77,1)
feature_test = feature_test.reshape(20,1)
```

```python
regressor.fit(feature_train,label_train)
```

```python
print(regressor.score(feature_test, label_test))
```

```python
pred = regressor.predict(feature_test)
```

```python
pd.DataFrame(zip(pred,label_test))
```

```python
ans = regressor.predict([[3.073]])
```

```python
print("Profit is in negative, so it will give loss to start the buisness in this Jaipur. Predicted loss can be "+str(round(ans[0]*-1,3))+" Mn")
```
